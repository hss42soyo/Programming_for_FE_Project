```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

orders_1000 = pd.read_csv("./1000_orders.csv")
orders_10000 = pd.read_csv("./10000_orders.csv")
orders_50000 = pd.read_csv("./50000_orders.csv")
```

```python
plt.hist(orders_1000['Latency'], alpha=0.5, label='1000 orders')
plt.hist(orders_10000['Latency'], alpha=0.5, label='10000 orders')
plt.hist(orders_50000['Latency'], alpha=0.5, label='50000 orders')
plt.xlabel('Latency')
plt.ylabel('Frequency')
plt.title('Latency Distribution')
plt.legend(loc='upper right')
plt.show()
```
